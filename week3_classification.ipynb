## Getting data ready for models

```python
# loading data
url = 'https://raw.githubusercontent.com/ov3ipo/datatalksclub_ml_course/refs/heads/main/dataset/WA_Fn-UseC_-Telco-Customer-Churn.csv'
df = pd.read_csv(url)
df.head()
```

```python
# incase we can't see all the columns using transpose dataframe instead
df.head().T
```

```python
# (optional)
# change data frame names into more easier to manipulate ones
df.columns = df.columns.str.lower().str.replace(' ', '_')
categoricals_cols = list(df.dtypes[df.dtypes == 'object'].index)
for c in categoricals_cols:
    df[c] = df[c].str.lower().str.replace(' ', '_')

df.head().T
```

```python
# getting data types
df.dtypes
```
